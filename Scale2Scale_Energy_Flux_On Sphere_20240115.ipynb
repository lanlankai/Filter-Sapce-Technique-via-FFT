# Diagnose the scale-to-scale energy flux

In the field of turbulence, Richardson (1922) hypothesize a kinetic energy cascade  that the kinetic energy is flowing from the large scale vortex to a small scale one, until the so-called fluid viscosity scale, where the kinetic energy converts to heat. Kolmogorov (1941) proposed his famous homogeneous and isotropic turbulence theory to put the cascade idea forward, where a famous $-5/3$ power law of the velocity spectrum is predicted. In the past few decades, the idea of cascade has been generalized to cope with different situations, and was widely accepted (Frisch,1995; Schmitt & Huang, 2016; Alaxiz & Bifferale, 2018; Zhou, 2020; Benzi & Toschi, 2023). However, we still need a good indictor to measure the scale-to-scale energy flux that quantifies the basic hypothesis of the cascade.

There exist three methodogies that derived from the Navier-Stokes equations. The first one is the classical third-order structure-functions (Komogorov,1941). However, to interprete the result, one needs <em>a proiori</em> knowledge of the system, which is often infeasible for a real problem.  The second one is the spectral representation, in which the Navier-Stokes equation is written in Fourier space. It thus requires a regular domain without data missing. The third one is the Filter-Space-Technique, which is from the field of Large-Eddy-Simulation (Lenard, 1971). It can provide the local information. The main shortcoming of this technique is its heavy calculation since a convolution is involved for many times. 



Filter-Space-Technique to perform a scale-to-scale flux analysis.
The Gassuian-kernel in 2D is written as, 
$$G_{\ell}(\mathbf{x})\propto \exp\left( -\frac{\vert \mathbf{x}\vert^2}{2{\ell}^2}\right)$$
where $\ell$ is the radius of the Gaussian kernel.
The coarse-grained operation is defined as,
$$u_{\ell}(x)=\int_{\vert x -x'\vert\le \ell} u(x+x')G_{\ell}(x') d x'$$
The energy flux can be analytically derived from the Navier-Stokes equaiton, which is written as,
$$\Pi^{\ell}(x)=-\left[ (u_iu_j)^{\ell}-u_i^{\ell}u_j^{\ell} \right]\frac{\partial u_i^{\ell}}{\partial x_j}$$
The scalar flux can be derived in the same way as, 
$$\Pi_{\phi}^{\ell}(x)=-\left[ (u_j\phi)^{\ell}-u_j^{\ell}\phi^{\ell} \right]\frac{\partial \phi^{\ell}}{\partial x_j}$$



## Coarse-graining on the sphere
For any scalar field $\psi(\mathbf{x})$, its coarse-grained field (resp. low-pass filtered), $\psi^{[\ell]}(\mathbf{x})$, by convolving $\psi(\mathbf{x})$ with a normalized filter kernel $G_{\ell}(\mathbf{r})$ is defined as,
$$\psi^{[\ell]}(\mathbf{x})=G_{\ell} \ast  \psi (\mathbf{x})$$
where $\ast$ is convolution on the sphere, $\mathbf{x}$ is geo-location on the globe; the kernel $G_{\ell}$ is often taken the Gaussian form, and $\ell$ is the radius of the kernel. 
The kernel is area normlized, 
$$\int G_{\ell}(\mathbf{X}) d\mathcal{S}=1$$
where $d\mathcal{S}$ is the area element on the sphere.

The above formalism holds for coarse-graining scalar fields. To coarse-grain a vector field on a sphere generally requires more work. Considering here the surface geostrophic velocity, $\mathbf{u}(\mathbf{x},t)=(v_{\lambda},v_{\phi})$ ($\lambda$ and $\phi$ are longitude and latitude) is divergence free on the 2D spherical surface. 

## exact convolution on sphere

For non-divergent vector fields, coarse-graining $\mathbf{u}$ is equivalent to coarse-graining each of its Cartesian components. The vector from spherical $(u_r,u_{\lambda},u_{\phi})$ to planetary Cartesian coordinates $(u_x,u_y,u_z)$ is then written as,
$$\begin{eqnarray*}
u_x&=&u_r\cos(\lambda) \cos(\phi)-u_{\lambda}\sin(\lambda)-u_{\phi}\cos(\lambda)\sin(\phi) \\
u_y&=&u_r\sin(\lambda) \cos(\phi)+u_{\lambda}\cos(\lambda)-u_{\phi}\sin(\lambda)\sin(\phi) \\
u_z&=&u_r\sin(\phi) +u_{\phi}\cos(\phi) \\
\end{eqnarray*}$$
where $\lambda$ is longitude and $\phi$ is latitude, respectively, and $u_{\lambda}$, $u_{\phi}$ are the zonal and meridional velocity components. 
The radial velocity component, $u_r=0$ for the geostrophic flow.  Taking the velocity components $(u_x,u_y,u_z)$ each as scalar field to obtain the corresponding coarse-grained fields $(\overline{u}_x,\overline{u}_y,\overline{u}_z)$, the coarse-grained velocity, $\mathbf{u}^{[\ell]}$ in the spherical coordinates is then written as
$$\begin{eqnarray*}
\overline{u}_r&=&\overline{u}_x\cos(\lambda) \cos(\phi)+\overline{u}_{y}\sin(\lambda)\cos(\phi)+\overline{u}_{z}\sin(\phi)=0 \\
\overline{u}_\lambda&=&-\overline{u}_x\sin(\lambda) +\overline{u}_{y}\cos(\lambda)\\
\overline{u}_\phi&=&-\overline{u}_x\cos(\lambda) \sin(\phi)-\overline{u}_{y}\sin(\lambda)\sin(\phi)+\overline{u}_{z}\cos(\phi)\\
\end{eqnarray*}$$
The spatial gradient of the Cartesian velocity is written as,
\begin{gather}
 \begin{bmatrix}  \partial_x \\   \partial_y \\    \partial_z \end{bmatrix}u
 = \begin{bmatrix}
 \cos(\lambda)\cos(\phi) &
-\frac{\sin(\lambda)}{r\cos(\phi)} &-\frac{\cos(\lambda)\sin(\phi)}{r} \\
    \sin(\lambda)\cos(\phi) &
\frac{\cos(\lambda)}{r\cos(\phi)} &-\frac{\sin(\lambda)\sin(\phi)}{r} \\
\sin(\phi) &
0 &\frac{\cos(\phi)}{r} \\
   \end{bmatrix}\begin{bmatrix}  \partial_r \\   \partial_\lambda \\    \partial_\phi \end{bmatrix}u
\end{gather}
In the case that we're restricing ourselves to a spherical shell, then $ \partial _r\equiv0 $, so this reduced down to
\begin{gather}
 \begin{bmatrix}  \partial_x \\   \partial_y \\    \partial_z \end{bmatrix}u
 = \begin{bmatrix}
-\frac{\sin(\lambda)}{r\cos(\phi)} &-\frac{\cos(\lambda)\sin(\phi)}{r} \\
\frac{\cos(\lambda)}{r\cos(\phi)} &-\frac{\sin(\lambda)\sin(\phi)}{r} \\
0 &\frac{\cos(\phi)}{r} \\
   \end{bmatrix}\begin{bmatrix} \partial_\lambda \\    \partial_\phi \end{bmatrix}u
\end{gather}

## define function

```python
def prepare_data_Sphere_2_Cartesian(
    da,
):  # OK convert the velocity from spheric coordinates to Cartesian coordinates.
    # u is the east-west velocity, and v is the south to north velocity
    # The radial velocity is assume to be zero. Note that the incompressible condition is satisfied in 3D.
    # Therefore, a 2D cut is compressible. One can still apply the partial scale-to-scale energy flux estimator as mentioned above.

    # convert degree to radian
    da["latitude"] = np.deg2rad(da["latitude"])
    da["longitude"] = np.deg2rad(da["longitude"])

    da["sin_lon"] = np.sin(da["longitude"])
    da["cos_lon"] = np.cos(da["longitude"])
    da["sin_lat"] = np.sin(da["latitude"])
    da["cos_lat"] = np.cos(da["latitude"])

    da["ux"] = -da["u"] * da["sin_lon"] - da["v"] * da["cos_lon"] * da["sin_lat"]
    da["ux"].attrs["long_name"] = "velocity component x in Cartesian coordinates"
    da["uy"] = da["u"] * da["cos_lon"] - da["v"] * da["sin_lon"] * da["sin_lat"]
    da["uy"].attrs["long_name"] = "velocity component y in Cartesian coordinates"

    da["uz"] = da["v"] * da["cos_lat"]

    da["uz"].attrs["long_name"] = "velocity component z in Cartesian coordinates"

    da["Earth_radius"] = 6378.137
    da["Earth_radius"].attrs["long_name"] = "Earth radius in km"

    # we assume an equal grid in degree on the sphere
    da["delta_lon"] = (da["longitude"][1] - da["longitude"][0]).values
    da["delta_lon"].attrs["long_name"] = "longitude resolution in radian"
    da["delta_lat"] = (da["latitude"][1] - da["latitude"][0]).values
    da["delta_lat"].attrs["long_name"] = "latitude resolution in radian"

    # area of each grid cell
    da["area"] = (
        da["Earth_radius"] ** 2
        * np.cos(da["latitude"])
        * da["delta_lon"]
        * da["delta_lat"]
    )
    # label land, island or missing data
    da["Flag"] = xr.DataArray(
        data=np.ones_like(da["u"]), dims=["latitude", "longitude"]
    )
    da["Flag"] = da["Flag"].where(da["u"].notnull(), 0)
    da["Flag"].attrs["long_name"] = "Flag for NaN"

    da.attrs[
        "version"
    ] = "Scale-to-Scale energy flux: Python code by Yongxiang Huang at Xiamen University China"

    # convert radian to degree
    da["latitude"] = np.rad2deg(da["latitude"])
    da["longitude"] = np.rad2deg(da["longitude"])

    return da


def Gaussian_kernel2d_Sphere(  # OK Gaussian kernel and velocity at given latitude
    da, ell, center_lat=0, center_lon=0, Num_factor=1.5
):
    """
    Generate the Gaussian kernel with an equal distrance to the center.
    Input:
        da['ux','uy','uz'] are the Cartesian velocity converted from the geostrophic flow (u zonal velocity,v meridional velocity)
        ell is the radius in km of the circle for coarse-graining (CG)
        center_lat is the centeral latitude for the Gaussian kernel
        center_lon is the centeral longitude for the Gaussian kernel
        Num_factor is the factor to choose the range to estimate the Gaussian kernel
    Output:
        da['kernel'] is the Gausian kernel for CG, it is normalzied by the area of each grid cell
        da['r'] is the distance
    """
    if ((center_lon >= 0) and (center_lon < 359)) or (
        (center_lon >= -180) and (center_lon < 179)
    ):
        r = great_circle(center_lat, center_lon, center_lat, center_lon + 1)
    else:
        r = great_circle(center_lat, center_lon, center_lat, center_lon - 1)

    delta_lon = ell / r * Num_factor

    if (center_lat >= -90) and (center_lat < 89):
        r = great_circle(center_lat, center_lon, center_lat + 1, center_lon)
    else:
        r = great_circle(center_lat, center_lon, center_lat - 1, center_lon)

    delta_lat = ell / r * Num_factor

    Index_lon = slice(center_lon - delta_lon, center_lon + delta_lon)
    Index_lat = slice(center_lat - delta_lat, center_lat + delta_lat)

    da_new = da.sel(longitude=Index_lon, latitude=Index_lat)
    grid_lon, grid_lat = np.meshgrid(da_new["longitude"], da_new["latitude"])

    dr = great_circle(center_lat, center_lon, grid_lat, grid_lon)

    da_mask = xr.DataArray(data=dr, dims=["latitude", "longitude"]).to_dataset(name="r")
    da_mask["r"].attrs["long_name"] = "distance from the kernel center in km"
    da_mask["area"] = da_new["area"]
    da_mask = da_mask.where(da_mask["r"] <= ell, drop=True)
    da_mask["kernel"] = np.exp(-da_mask["r"] ** 2 / (2 * ell**2)) * da_mask["area"]
    da_mask["kernel"] = da_mask["kernel"] / da_mask["kernel"].sum()
    da_mask["kernel"].attrs[
        "long_name"
    ] = "normalized Gaussian kernel by weighted of cell area"
    da_mask = da_mask.fillna(0)
    da_mask = da_mask.rename({"latitude": "y", "longitude": "x"})
    da["kernel"] = da_mask["kernel"]
    da["r"] = da_mask["r"]
    return da


def convolve_xarray(x, kernel, periodic=False):
    # periodic extension in the longitude direction for global analysis
    if periodic:
        Num_shape = np.shape(kernel)[0] // 2
        latitude = x["latitude"].values
        longitude_old = x["longitude"].values
        x = x.values
        x_new = np.hstack([x[:, -Num_shape:], x, x[:, :Num_shape]])
        longitude_new = np.hstack(
            [longitude_old[-Num_shape:], longitude_old, longitude_old[:Num_shape]]
        )
        x_new = convolve(x_new, kernel, mode="same")
        return xr.DataArray(
            data=x_new[:, Num_shape:-Num_shape], dims=["latitude", "longitude"]
        )

    else:
        return xr.DataArray(
            data=convolve(x.values, kernel, mode="same"), dims=["latitude", "longitude"]
        )


def coarse_grain_Sphere(  # coarse_graining at single latitude
    da,
    periodic=False,
):
    """
    Coarse-graining with a given kernel on Earth sphere
    Input:
        da contains the Cartesian velocity components and the Flag from prepare_data_Sphere_2_Cartesian
        and kernel information from Gaussian_kernel2d_Sphere
    Output:
        da_CG are the filtered velocity and the mask for a given latitude
    """

    # ux
    da["ux_cg"] = convolve_xarray(da["ux"], da["kernel"], periodic=periodic)
    da["uxux_cg"] = convolve_xarray(
        da["ux"] * da["ux"], da["kernel"], periodic=periodic
    )
    da["uxuy_cg"] = convolve_xarray(
        da["ux"] * da["uy"], da["kernel"], periodic=periodic
    )
    da["uxuz_cg"] = convolve_xarray(
        da["ux"] * da["uz"], da["kernel"], periodic=periodic
    )

    # uy
    da["uy_cg"] = convolve_xarray(da["uy"], da["kernel"], periodic=periodic)
    da["uyuy_cg"] = convolve_xarray(
        da["uy"] * da["uy"], da["kernel"], periodic=periodic
    )
    da["uyuz_cg"] = convolve_xarray(
        da["uy"] * da["uz"], da["kernel"], periodic=periodic
    )

    # z
    da["uz_cg"] = convolve_xarray(da["uz"], da["kernel"], periodic=periodic)
    da["uzuz_cg"] = convolve_xarray(
        da["uz"] * da["uz"], da["kernel"], periodic=periodic
    )

    # mask
    da["mask"] = convolve_xarray(da["Flag"], da["kernel"], periodic=periodic)
    da_CG = da[
        [
            "ux_cg",
            "uy_cg",
            "uz_cg",
            "uxux_cg",
            "uxuy_cg",
            "uxuz_cg",
            "uyuy_cg",
            "uyuz_cg",
            "uzuz_cg",
            "sin_lon",
            "cos_lon",
            "sin_lat",
            "cos_lat",
            "Earth_radius",
            "delta_lon",
            "delta_lat",
            "mask",
        ]
    ]

    return da_CG.copy()


def Energy_Flux_Aluie_Sphere(da_CG):
    da_CG["latitude"] = np.deg2rad(da_CG["latitude"])
    da_CG["longitude"] = np.deg2rad(da_CG["longitude"])

    A11 = -da_CG["sin_lon"] / da_CG["cos_lat"] / da_CG["Earth_radius"]
    A12 = -da_CG["cos_lon"] * da_CG["sin_lat"] / da_CG["Earth_radius"]
    A21 = da_CG["cos_lon"] / da_CG["cos_lat"] / da_CG["Earth_radius"]
    A22 = -da_CG["sin_lon"] * da_CG["sin_lat"] / da_CG["Earth_radius"]
    A32 = da_CG["cos_lat"] / da_CG["Earth_radius"]
    # ux-gradient
    ux_cg_lon = da_CG["ux_cg"].differentiate("longitude")
    ux_cg_lat = da_CG["ux_cg"].differentiate("latitude")
    da_CG["ux_cg_x"] = A11 * ux_cg_lon + A12 * ux_cg_lat
    da_CG["ux_cg_y"] = A21 * ux_cg_lon + A22 * ux_cg_lat
    da_CG["ux_cg_z"] = A32 * ux_cg_lat

    # uy-gradient
    uy_cg_lon = da_CG["uy_cg"].differentiate("longitude")
    uy_cg_lat = da_CG["uy_cg"].differentiate("latitude")
    da_CG["uy_cg_x"] = A11 * uy_cg_lon + A12 * uy_cg_lat
    da_CG["uy_cg_y"] = A21 * uy_cg_lon + A22 * uy_cg_lat
    da_CG["uy_cg_z"] = A32 * uy_cg_lat

    # uy-gradient
    uz_cg_lon = da_CG["uz_cg"].differentiate("longitude")
    uz_cg_lat = da_CG["uz_cg"].differentiate("latitude")
    da_CG["uz_cg_x"] = A11 * uz_cg_lon + A12 * uz_cg_lat
    da_CG["uz_cg_y"] = A21 * uz_cg_lon + A22 * uz_cg_lat
    da_CG["uz_cg_z"] = A32 * uz_cg_lat

    # ux
    da_CG["EF"] = (da_CG["ux_cg"] * da_CG["ux_cg"] - da_CG["uxux_cg"]) * da_CG[
        "ux_cg_x"
    ]
    da_CG["EF"] += (da_CG["ux_cg"] * da_CG["uy_cg"] - da_CG["uxuy_cg"]) * da_CG[
        "ux_cg_y"
    ]
    da_CG["EF"] += (da_CG["ux_cg"] * da_CG["uz_cg"] - da_CG["uxuz_cg"]) * da_CG[
        "ux_cg_z"
    ]

    # uy
    da_CG["EF"] += (da_CG["uy_cg"] * da_CG["ux_cg"] - da_CG["uxuy_cg"]) * da_CG[
        "uy_cg_x"
    ]
    da_CG["EF"] += (da_CG["uy_cg"] * da_CG["uy_cg"] - da_CG["uyuy_cg"]) * da_CG[
        "uy_cg_y"
    ]
    da_CG["EF"] += (da_CG["uy_cg"] * da_CG["uz_cg"] - da_CG["uyuz_cg"]) * da_CG[
        "uy_cg_z"
    ]

    # uz
    da_CG["EF"] += (da_CG["uz_cg"] * da_CG["ux_cg"] - da_CG["uxuz_cg"]) * da_CG[
        "uz_cg_x"
    ]
    da_CG["EF"] += (da_CG["uz_cg"] * da_CG["uy_cg"] - da_CG["uyuz_cg"]) * da_CG[
        "uz_cg_y"
    ]
    da_CG["EF"] += (da_CG["uz_cg"] * da_CG["uz_cg"] - da_CG["uzuz_cg"]) * da_CG[
        "uz_cg_z"
    ]

    da_CG["latitude"] = np.rad2deg(da_CG["latitude"])
    da_CG["longitude"] = np.rad2deg(da_CG["longitude"])

    return da_CG[["EF", "mask"]].copy()
```
